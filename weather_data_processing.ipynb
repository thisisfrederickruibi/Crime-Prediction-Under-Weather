```python
import pandas as pd
import os
```

```python
# Set folder path relative to notebook
folder = "Greenwich"
```

```python
# Load CSVs from inside the folder
df1 = pd.read_csv(os.path.join(folder, "Greenwich_1.csv"))
df2 = pd.read_csv(os.path.join(folder, "Greenwich_2.csv"))
#df3 = pd.read_csv(os.path.join(folder, "Islington_3.csv"))
```

```python
# Merge them into one DataFrame
merged_df = pd.concat([df1, df2], ignore_index=True)
```

```python
# Convert 'datetime' to proper datetime format and sort
merged_df['datetime'] = pd.to_datetime(merged_df['datetime'])
merged_df = merged_df.sort_values('datetime').reset_index(drop=True)
```

```python
# Preview the result
merged_df.head(10)
```

```python
# Select only the desired columns
columns_to_keep = [
    "name", "datetime", "tempmax", "tempmin", "temp",
    "humidity", "precip", "snow", "windspeed", "visibility"
]
```

```python
filtered_df = merged_df[columns_to_keep]
```

```python
# Preview the result
filtered_df.head(10)
```

```python
# Set all values in the 'name' column to borough name
filtered_df['name'] = "Greenwich"
```

```python
# Preview the result
filtered_df.head(10)
```

```python
# Check for duplicates in the datetime column
duplicate_dates = filtered_df[filtered_df.duplicated(subset='datetime')]

print("Duplicate dates:")
print(duplicate_dates)
```

```python
# Create a complete date range from min to max date
full_range = pd.date_range(start=filtered_df['datetime'].min(),
                           end=filtered_df['datetime'].max(),
                           freq='D')

# Find missing dates
existing_dates = pd.to_datetime(filtered_df['datetime'].unique())
missing_dates = full_range.difference(existing_dates)

print(f"Missing dates ({len(missing_dates)}):")
print(missing_dates)
```

```python
clean_df = filtered_df.drop_duplicates(subset='datetime', keep='first')
```

```python
# Preview the result
clean_df.head(10)
```

```python
# Check for duplicates in the datetime column
duplicate_dates = clean_df[filtered_df.duplicated(subset='datetime')]

print("Duplicate dates:")
print(duplicate_dates)
```

```python
filtered_df.to_csv("Greenwich.csv", index=False)
```
